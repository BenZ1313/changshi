```python
# based on tensorflow keras stock price prediction model comparison analysis

# ================================
# import necessary libraries and modules
# ================================

# standard library
import os  # used for interacting with the operating system
import json  # used for handling json data
import random  # used for generating random numbers
import math  # used for mathematical operations such as calculating square roots
import time  # import time module

# data processing and visualization
import numpy as np  # import numpy library for efficient numerical computations
import pandas as pd  # import pandas library for data processing and analysis
import matplotlib.pyplot as plt  # import matplotlib library for plotting graphs

# deep learning and neural networks
import tensorflow as tf  # import tensorflow library for deep learning tasks
from tensorflow.keras.models import Sequential, load_model  # import keras sequential model and model loading functionality
from tensorflow.keras.layers import Dense, LSTM, Dropout, GRU, Bidirectional  # import commonly used keras layers
from tensorflow.keras.callbacks import EarlyStopping  # import early stopping callback

# preprocessing and evaluation
from sklearn.preprocessing import MinMaxScaler  # import minmaxscaler for feature normalization
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # import evaluation metrics
from sklearn.model_selection import train_test_split  # import train_test_split for splitting dataset

# ignore unnecessary warnings
import warnings
warnings.filterwarnings("ignore")  # ignore all warnings

# ================================
# set random seed to ensure experiment reproducibility
# ================================

seed_num = 42  # set random seed number
os.environ['PYTHONHASHSEED'] = str(seed_num)  # set python's hash seed
random.seed(seed_num)  # set python random seed
np.random.seed(seed_num)  # set numpy random seed
tf.random.set_seed(seed_num)  # set tensorflow random seed

# ================================
# set early stopping callback
# ================================

earlystop = EarlyStopping(
    monitor='val_loss',  # monitor the loss on the validation set
    min_delta=0,  # minimum change, changes less than this are considered no improvement
    patience=20,  # if validation loss does not improve within 20 epochs, stop training
    verbose=2  # level of detail in output during training
)

# ================================
# read and preprocess data
# ================================

# read stock data
data = pd.read_csv('IBM.csv')
print(data.head())  # output first few rows to check data format

# select required feature columns (excluding the date column)
all_data = data.iloc[:, 1:7]  # select columns 1 to 6 (excluding column 0 date)
print(all_data.head())  # display first few rows of the selected data

# further select specific columns
all_data = all_data.loc[:, ['Open', 'Close', 'High', 'Low', 'Volume', 'Adj Close']]
print(all_data.info())  # output basic information of the data
print(all_data.describe())  # output statistical information of the data

# plot histograms of the distribution for each feature
all_data.hist(bins=50, figsize=(20, 15))
plt.show()

# feature normalization, scaling all features to the range [0, 1]
sc = MinMaxScaler(feature_range=(0, 1))
all_data_scaled = sc.fit_transform(all_data)
print(all_data_scaled)  # output normalized data
print(':', len(all_data_scaled))  # output length of the data

# set time step to 60
features = []  # store features
labels = []  # store labels
for i in range(60, len(all_data_scaled)):  # start from the 60th data point
    features.append(all_data_scaled[i-60:i, ])  # take the previous 60 time steps as features
    labels.append(all_data_scaled[i, 1])  # use the 1st column 'close' as the label
features, labels = np.array(features), np.array(labels)  # convert to numpy arrays
features = np.reshape(features, (features.shape[0], features.shape[1], -1))  # reshape features to fit lstm input

# split the dataset into training, validation, and testing sets
x_train, x_val, x_test = features[:6550], features[6550:6914], features[6914:]
y_train, y_val, y_test = labels[:6550], labels[6550:6914], labels[6914:]

# print the shapes of the datasets
print('shape of x_train:', x_train.shape)
print('shape of x_val:', x_val.shape)
print('shape of x_test:', x_test.shape)
print('shape of y_train:', y_train.shape)
print('shape of y_val:', y_val.shape)
print('shape of y_test:', y_test.shape)
# ================================
# build and train lstm model
# ================================
lstart_time = time.time()  # record start time
# initialize lstm model
regressor = Sequential()

# add first lstm layer and dropout
regressor.add(LSTM(
    units=50, 
    return_sequences=True, 
    input_shape=(x_train.shape[1], 6)  # input shape is (time steps, number of features)
))
regressor.add(Dropout(0.2))  # prevent overfitting by dropping 20% of neurons

# add second lstm layer and dropout
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

# add third lstm layer and dropout
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

# add fourth lstm layer and dropout
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))

# add output layer
regressor.add(Dense(units=1))  # output a value: the predicted stock price

# compile lstm model
regressor.compile(optimizer='adam', loss='mean_squared_error')
print(regressor.summary())  # output model summary

# train lstm model
history = regressor.fit(
    x_train, y_train, 
    epochs=25,  # number of epochs, tested with 5 / 15 / 25
    batch_size=32,  # batch size of 32 samples
    validation_data=(x_val, y_val), 
    callbacks=[earlystop]  # using early stopping callback only
)

lend_time = time.time()  # record end time
lstm_runtime = lend_time - lstart_time

# plot training loss and validation loss curves for lstm
loss = history.history['loss']  # training loss
val_loss = history.history['val_loss']  # validation loss
epochs_range = range(1, len(loss) + 1)

plt.figure(figsize=(5, 3))
plt.title('lstm loss curve')
plt.plot(epochs_range, loss, 'red', label='training loss')
plt.plot(epochs_range, val_loss, 'blue', label='validation loss')
plt.grid(True)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()

# normalize target variable 'close' column
sc_one = MinMaxScaler(feature_range=(0, 1))
sc_one.fit(all_data.iloc[:, 1:2])  # only fit, do not transform

# predict stock prices on training and testing sets
predicted_stock_train = regressor.predict(x_train)
predicted_stock_train = sc_one.inverse_transform(predicted_stock_train)

predicted_stock_test_lstm = regressor.predict(x_test)
predicted_stock_test_lstm = sc_one.inverse_transform(predicted_stock_test_lstm)

# get the actual stock prices
real_price_train = sc_one.inverse_transform(np.reshape(y_train, (-1, 1)))
real_price_test = sc_one.inverse_transform(np.reshape(y_test, (-1, 1)))

# visualize lstm model predictions on the training set
plt.figure(figsize=(5, 3))
plt.plot(real_price_train, color='red', label='real stock price')
plt.plot(predicted_stock_train, color='blue', label='predicted stock price')
plt.title('lstm train stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.legend()
plt.grid(True)
plt.show()

# visualize lstm model predictions on the testing set
plt.figure(figsize=(5, 3))
plt.plot(real_price_test, color='red', label='real stock price')
plt.plot(predicted_stock_test_lstm, color='blue', label='predicted stock price')
plt.title('lstm test stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.legend()
plt.grid(True)
plt.show()

# ================================
# evaluate lstm model
# ================================

# calculate evaluation metrics
lstm_mse = mean_squared_error(predicted_stock_test_lstm, real_price_test)  # mean squared error
lstm_rmse = math.sqrt(lstm_mse)  # root mean squared error
lstm_mae = mean_absolute_error(predicted_stock_test_lstm, real_price_test)  # mean absolute error
lstm_r2 = r2_score(predicted_stock_test_lstm, real_price_test)  # r squared score

# output evaluation results
print('lstm model mse: %.6f' % lstm_mse)
print('lstm model rmse: %.6f' % lstm_rmse)
print('lstm model mae: %.6f' % lstm_mae)
print('lstm model r-squared: %.6f' % lstm_r2)
print('lstm model runtime: %.2f seconds' % lstm_runtime)
# ================================
# build and train gru model
# ================================
gstart_time = time.time()  # record start time
# initialize gru model
regressorGRU = Sequential()

# add first gru layer and dropout
regressorGRU.add(GRU(
    units=50, 
    return_sequences=True, 
    input_shape=(x_train.shape[1], 6)
))
regressorGRU.add(Dropout(0.2))

# add second gru layer and dropout
regressorGRU.add(GRU(units=50, return_sequences=True))
regressorGRU.add(Dropout(0.2))

# add third gru layer and dropout
regressorGRU.add(GRU(units=50, return_sequences=True))
regressorGRU.add(Dropout(0.2))

# add fourth gru layer and dropout
regressorGRU.add(GRU(units=50))
regressorGRU.add(Dropout(0.2))

# add output layer
regressorGRU.add(Dense(units=1))

# compile gru model
regressorGRU.compile(optimizer='adam', loss='mean_squared_error')
print(regressorGRU.summary())

# train gru model
history = regressorGRU.fit(
    x_train, y_train, 
    epochs=25, 
    batch_size=32, 
    validation_data=(x_val, y_val), 
    callbacks=[earlystop]  # using early stopping callback only
)

gend_time = time.time()  # record end time
gru_runtime = gend_time - gstart_time

# plot training and validation loss curves for gru
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(loss) + 1)

plt.figure(figsize=(5, 3))
plt.title('gru loss curve')
plt.plot(epochs_range, loss, 'red', label='training loss')
plt.plot(epochs_range, val_loss, 'blue', label='validation loss')
plt.grid(True)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()

# normalize target variable 'close' column (repeated operation to ensure consistency)
sc_one = MinMaxScaler(feature_range=(0, 1))
sc_one.fit(all_data.iloc[:, 1:2])  # only fit, do not transform

# predict stock prices on training and testing sets
predicted_stock_train = regressorGRU.predict(x_train)
predicted_stock_train = sc_one.inverse_transform(predicted_stock_train)

predicted_stock_test_gru = regressorGRU.predict(x_test)
predicted_stock_test_gru = sc_one.inverse_transform(predicted_stock_test_gru)

# get the actual stock prices
real_price_train = sc_one.inverse_transform(np.reshape(y_train, (-1, 1)))
real_price_test = sc_one.inverse_transform(np.reshape(y_test, (-1, 1)))

# visualize gru model predictions on the training set
plt.figure(figsize=(5, 3))
plt.plot(real_price_train, color='red', label='real stock price')
plt.plot(predicted_stock_train, color='blue', label='predicted stock price')
plt.title('gru train stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# visualize gru model predictions on the testing set
plt.figure(figsize=(5, 3))
plt.plot(real_price_test, color='red', label='real stock price')
plt.plot(predicted_stock_test_gru, color='blue', label='predicted stock price')
plt.title('gru test stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# ================================
# evaluate gru model
# ================================

# calculate evaluation metrics
gru_mse = mean_squared_error(predicted_stock_test_gru, real_price_test)
gru_rmse = math.sqrt(gru_mse)
gru_mae = mean_absolute_error(predicted_stock_test_gru, real_price_test)
gru_r2 = r2_score(predicted_stock_test_gru, real_price_test)

# output evaluation results
print('gru model mse: %.6f' % gru_mse)
print('gru model rmse: %.6f' % gru_rmse)
print('gru model mae: %.6f' % gru_mae)
print('gru model r-squared: %.6f' % gru_r2)
print('gru model runtime: %.2f seconds' % gru_runtime)
# ================================
# build and train bilstm model
# ================================
bistart_time = time.time()  # record start time
# initialize bilstm model
regressorBiLSTM = Sequential()

# add first bidirectional lstm layer and dropout
regressorBiLSTM.add(Bidirectional(
    LSTM(
        units=50, 
        return_sequences=True
    ), 
    input_shape=(x_train.shape[1], 6), 
    merge_mode='concat'  # concatenate forward and backward outputs
))
regressorBiLSTM.add(Dropout(0.2))

# add second bidirectional lstm layer and dropout
regressorBiLSTM.add(Bidirectional(LSTM(units=50, return_sequences=True)))
regressorBiLSTM.add(Dropout(0.2))

# add third bidirectional lstm layer and dropout
regressorBiLSTM.add(Bidirectional(LSTM(units=50, return_sequences=True)))
regressorBiLSTM.add(Dropout(0.2))

# add fourth bidirectional lstm layer and dropout
regressorBiLSTM.add(Bidirectional(LSTM(units=50)))
regressorBiLSTM.add(Dropout(0.2))

# add output layer
regressorBiLSTM.add(Dense(units=1))  # output one predicted value

# compile bilstm model
regressorBiLSTM.compile(optimizer='adam', loss='mean_squared_error')
print(regressorBiLSTM.summary())

# train bilstm model
history = regressorBiLSTM.fit(
    x_train, y_train, 
    epochs=25, 
    batch_size=32, 
    validation_data=(x_val, y_val), 
    callbacks=[earlystop]  # using early stopping callback only
)

biend_time = time.time()  # record end time
bi_runtime = biend_time - bistart_time

# plot training and validation loss curves for bilstm
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(loss) + 1)

plt.figure(figsize=(5, 3))
plt.title('bilstm loss curve')
plt.plot(epochs_range, loss, 'red', label='training loss')
plt.plot(epochs_range, val_loss, 'blue', label='validation loss')
plt.grid(True)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()

# normalize target variable 'close' column (repeated operation to ensure consistency)
sc_one = MinMaxScaler(feature_range=(0, 1))
sc_one.fit(all_data.iloc[:, 1:2])  # only fit, do not transform

# predict stock prices on training and testing sets
predicted_stock_train = regressorBiLSTM.predict(x_train)
predicted_stock_train = sc_one.inverse_transform(predicted_stock_train)

predicted_stock_test_bilstm = regressorBiLSTM.predict(x_test)
predicted_stock_test_bilstm = sc_one.inverse_transform(predicted_stock_test_bilstm)

# get the actual stock prices
real_price_train = sc_one.inverse_transform(np.reshape(y_train, (-1, 1)))
real_price_test = sc_one.inverse_transform(np.reshape(y_test, (-1, 1)))

# visualize bilstm model predictions on the training set
plt.figure(figsize=(5, 3))
plt.plot(real_price_train, color='red', label='real stock price')
plt.plot(predicted_stock_train, color='blue', label='predicted stock price')
plt.title('bilstm train stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# visualize bilstm model predictions on the testing set
plt.figure(figsize=(5, 3))
plt.plot(real_price_test, color='red', label='real stock price')
plt.plot(predicted_stock_test_bilstm, color='blue', label='predicted stock price')
plt.title('bilstm test stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# ================================
# evaluate bilstm model
# ================================

# calculate evaluation metrics
bi_mse = mean_squared_error(predicted_stock_test_bilstm, real_price_test)
bi_rmse = math.sqrt(bi_mse)
bi_mae = mean_absolute_error(predicted_stock_test_bilstm, real_price_test)
bi_r2 = r2_score(predicted_stock_test_bilstm, real_price_test)

# output evaluation results
print('bilstm model mse: %.6f' % bi_mse)
print('bilstm model rmse: %.6f' % bi_rmse)
print('bilstm model mae: %.6f' % bi_mae)
print('bilstm model r-squared: %.6f' % bi_r2)
print('bilstm model runtime: %.2f seconds' % bi_runtime)
# ================================
# baseline models
# ================================

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# ================================
# prepare data: convert to 2d format
# ================================
# convert features to 2d format
features_2d = np.reshape(features, (features.shape[0], -1))

# split the 2d dataset into training, validation, and testing sets
x_train_2d, x_val_2d, x_test_2d = features_2d[:6550], features_2d[6550:6914], features_2d[6914:]
y_train, y_val, y_test = labels[:6550], labels[6550:6914], labels[6914:]

# ================================
# build and train linear regression model
# ================================
lr_model = LinearRegression()

lrstart_time = time.time()  # record start time
lr_model.fit(x_train_2d, y_train)
lrend_time = time.time()  # record end time

lr_runtime = lrend_time - lrstart_time

# predictions
lr_train_pred = lr_model.predict(x_train_2d)
lr_test_pred = lr_model.predict(x_test_2d)

# inverse transform predicted values
lr_train_pred_inv = sc_one.inverse_transform(lr_train_pred.reshape(-1, 1))
lr_test_pred_inv = sc_one.inverse_transform(lr_test_pred.reshape(-1, 1))

# inverse transform actual values
real_price_train = sc_one.inverse_transform(np.reshape(y_train, (-1, 1)))
real_price_test = sc_one.inverse_transform(np.reshape(y_test, (-1, 1)))

# visualize linear regression predictions on the testing set
plt.figure(figsize=(5, 3))
plt.plot(real_price_test, color='red', label='real stock price')
plt.plot(lr_test_pred_inv, color='blue', label='predicted stock price')
plt.title('linear regression test stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# ================================
# build and train random forest model
# ================================
rf_model = RandomForestRegressor(n_estimators=100, random_state=seed_num)

rfstart_time = time.time()  # record start time
rf_model.fit(x_train_2d, y_train)
rfend_time = time.time()  # record end time
rf_runtime = rfend_time - rfstart_time

# predictions
rf_train_pred = rf_model.predict(x_train_2d)
rf_test_pred = rf_model.predict(x_test_2d)

# inverse transform predicted values
rf_train_pred_inv = sc_one.inverse_transform(rf_train_pred.reshape(-1, 1))
rf_test_pred_inv = sc_one.inverse_transform(rf_test_pred.reshape(-1, 1))

# visualize random forest predictions on the testing set
plt.figure(figsize=(5, 3))
plt.plot(real_price_test, color='red', label='real stock price')
plt.plot(rf_test_pred_inv, color='blue', label='predicted stock price')
plt.title('random forest test stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.grid(True)
plt.legend()
plt.show()

# ================================
# evaluate linear regression and random forest models
# ================================
# linear regression evaluation
lr_mse = mean_squared_error(real_price_test, lr_test_pred_inv)
lr_rmse = math.sqrt(lr_mse)
lr_mae = mean_absolute_error(real_price_test, lr_test_pred_inv)
lr_r2 = r2_score(real_price_test, lr_test_pred_inv)

print('linear regression model mse: %.6f' % lr_mse)
print('linear regression model rmse: %.6f' % lr_rmse)
print('linear regression model mae: %.6f' % lr_mae)
print('linear regression model r-squared: %.6f' % lr_r2)
print('linear regression model runtime: %.2f seconds' % lr_runtime)

# random forest evaluation
rf_mse = mean_squared_error(real_price_test, rf_test_pred_inv)
rf_rmse = math.sqrt(rf_mse)
rf_mae = mean_absolute_error(real_price_test, rf_test_pred_inv)
rf_r2 = r2_score(real_price_test, rf_test_pred_inv)

print('random forest model mse: %.6f' % rf_mse)
print('random forest model rmse: %.6f' % rf_rmse)
print('random forest model mae: %.6f' % rf_mae)
print('random forest model r-squared: %.6f' % rf_r2)
print('random forest model runtime: %.2f seconds' % rf_runtime)
# ================================
# plot comparison charts
# ================================

# evaluation data for models
metrics_values = {
    'lstm': {'mse': lstm_mse, 'rmse': lstm_rmse, 'mae': lstm_mae, 'r2': lstm_r2, 'time': lstm_runtime},
    'gru': {'mse': gru_mse, 'rmse': gru_rmse, 'mae': gru_mae, 'r2': gru_r2, 'time': gru_runtime},
    'bilstm': {'mse': bi_mse, 'rmse': bi_rmse, 'mae': bi_mae, 'r2': bi_r2, 'time': bi_runtime},
    'lr': {'mse': lr_mse, 'rmse': lr_rmse, 'mae': lr_mae, 'r2': lr_r2, 'time': lr_runtime},
    'rf': {'mse': rf_mse, 'rmse': rf_rmse, 'mae': rf_mae, 'r2': rf_r2, 'time': rf_runtime}
}

# extract data for plotting charts
metrics = ['mse', 'rmse', 'mae', 'r2', 'time']
models = list(metrics_values.keys())

# plot bar charts for each metric
for metric in metrics:
    metric_scores = [metrics_values[model][metric] for model in models]
    
    plt.figure(figsize=(5, 3))
    plt.bar(models, metric_scores)
    plt.title(f'{metric} comparison across models')
    plt.xlabel('models')
    plt.ylabel(metric)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# ================================
# statistical tests: t-test and f-test
# ================================

# import additional libraries
from scipy.stats import ttest_rel, levene
import matplotlib.pyplot as plt
import seaborn as sns

# ================================
# calculate residuals
# ================================

# lstm residuals
residuals_lstm = real_price_test - predicted_stock_test_lstm

# gru residuals
residuals_gru = real_price_test - predicted_stock_test_gru

# bilstm residuals
residuals_bilstm = real_price_test - predicted_stock_test_bilstm

# linear regression residuals
residuals_lr = real_price_test - lr_test_pred_inv

# random forest residuals
residuals_rf = real_price_test - rf_test_pred_inv

# ================================
# perform t-test
# ================================

print('===============================')
print('t-test results:')
print('===============================')

# create a list to store t-test results
t_test_results = []

# lstm vs gru
t_stat, p_value = ttest_rel(residuals_lstm.flatten(), residuals_gru.flatten())
print('t-test (lstm vs gru):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'lstm vs gru', 't_stat': t_stat, 'p_value': p_value})

# lstm vs bilstm
t_stat, p_value = ttest_rel(residuals_lstm.flatten(), residuals_bilstm.flatten())
print('t-test (lstm vs bilstm):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'lstm vs bilstm', 't_stat': t_stat, 'p_value': p_value})

# lstm vs linear regression
t_stat, p_value = ttest_rel(residuals_lstm.flatten(), residuals_lr.flatten())
print('t-test (lstm vs linear regression):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'lstm vs linear regression', 't_stat': t_stat, 'p_value': p_value})

# lstm vs random forest
t_stat, p_value = ttest_rel(residuals_lstm.flatten(), residuals_rf.flatten())
print('t-test (lstm vs random forest):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'lstm vs random forest', 't_stat': t_stat, 'p_value': p_value})

# gru vs bilstm
t_stat, p_value = ttest_rel(residuals_gru.flatten(), residuals_bilstm.flatten())
print('t-test (gru vs bilstm):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'gru vs bilstm', 't_stat': t_stat, 'p_value': p_value})

# linear regression vs random forest
t_stat, p_value = ttest_rel(residuals_lr.flatten(), residuals_rf.flatten())
print('t-test (linear regression vs random forest):')
print('t statistic = %.4f, p value = %.4f' % (t_stat, p_value))
t_test_results.append({'model pair': 'linear regression vs random forest', 't_stat': t_stat, 'p_value': p_value})

# convert to dataframe
t_test_df = pd.DataFrame(t_test_results)


# ================================
# perform f-test (levene's test)
# ================================

print('===============================')
print('f-test (levenes test) results:')
print('===============================')

# create a list to store f-test results
f_test_results = []

# lstm vs gru
f_stat, p_value = levene(residuals_lstm.flatten(), residuals_gru.flatten())
print('f-test (lstm vs gru):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'lstm vs gru', 'f_stat': f_stat, 'p_value': p_value})

# lstm vs bilstm
f_stat, p_value = levene(residuals_lstm.flatten(), residuals_bilstm.flatten())
print('f-test (lstm vs bilstm):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'lstm vs bilstm', 'f_stat': f_stat, 'p_value': p_value})

# lstm vs linear regression
f_stat, p_value = levene(residuals_lstm.flatten(), residuals_lr.flatten())
print('f-test (lstm vs linear regression):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'lstm vs linear regression', 'f_stat': f_stat, 'p_value': p_value})

# lstm vs random forest
f_stat, p_value = levene(residuals_lstm.flatten(), residuals_rf.flatten())
print('f-test (lstm vs random forest):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'lstm vs random forest', 'f_stat': f_stat, 'p_value': p_value})

# gru vs bilstm
f_stat, p_value = levene(residuals_gru.flatten(), residuals_bilstm.flatten())
print('f-test (gru vs bilstm):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'gru vs bilstm', 'f_stat': f_stat, 'p_value': p_value})

# linear regression vs random forest
f_stat, p_value = levene(residuals_lr.flatten(), residuals_rf.flatten())
print('f-test (linear regression vs random forest):')
print('f statistic = %.4f, p value = %.4f' % (f_stat, p_value))
f_test_results.append({'model pair': 'linear regression vs random forest', 'f_stat': f_stat, 'p_value': p_value})

# convert to dataframe
f_test_df = pd.DataFrame(f_test_results)


# ================================
# visualize residual distributions
# ================================

# create a dataframe containing all residuals
residuals_df = pd.DataFrame({
    'lstm': residuals_lstm.flatten(),
    'gru': residuals_gru.flatten(),
    'bilstm': residuals_bilstm.flatten(),
    'linear regression': residuals_lr.flatten(),
    'random forest': residuals_rf.flatten()
})

# plot boxplot to compare residual distributions of different models
plt.figure(figsize=(12, 6))
sns.boxplot(data=residuals_df, palette='Set3')
plt.title('box plot of residual distributions for different models')
plt.xlabel('models')
plt.ylabel('residual distributions')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# plot violin plot to compare residual distributions of different models
plt.figure(figsize=(12, 6))
sns.violinplot(data=residuals_df, palette='Set2', inner='quartile')
plt.title('violin plot of residual distributions for different models')
plt.xlabel('models')
plt.ylabel('residual distributions')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# ================================
# cohen's d effect size comparison for all model residuals
# ================================

# define cohen's d function
def cohen_d(x, y):
    diff = x - y
    return np.mean(diff) / np.std(diff, ddof=1)

# calculate effect sizes between all model pairs
d_lstm_gru = cohen_d(residuals_lstm.flatten(), residuals_gru.flatten())
d_lstm_bilstm = cohen_d(residuals_lstm.flatten(), residuals_bilstm.flatten())
d_lstm_lr = cohen_d(residuals_lstm.flatten(), residuals_lr.flatten())
d_lstm_rf = cohen_d(residuals_lstm.flatten(), residuals_rf.flatten())
d_gru_bilstm = cohen_d(residuals_gru.flatten(), residuals_bilstm.flatten())
d_lr_rf = cohen_d(residuals_lr.flatten(), residuals_rf.flatten())

# print results
print(f"cohen's d for lstm vs gru: {d_lstm_gru:.4f}")
print(f"cohen's d for lstm vs bilstm: {d_lstm_bilstm:.4f}")
print(f"cohen's d for lstm vs linear regression: {d_lstm_lr:.4f}")
print(f"cohen's d for lstm vs random forest: {d_lstm_rf:.4f}")
print(f"cohen's d for gru vs bilstm: {d_gru_bilstm:.4f}")
print(f"cohen's d for linear regression vs random forest: {d_lr_rf:.4f}")

# summarize cohen's d results
models = [
    "lstm vs gru",
    "lstm vs bilstm",
    "lstm vs linear regression",
    "lstm vs random forest",
    "gru vs bilstm",
    "linear regression vs random forest"
]
cohen_d_values = [d_lstm_gru, d_lstm_bilstm, d_lstm_lr, d_lstm_rf, d_gru_bilstm, d_lr_rf]

# plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, cohen_d_values, color='skyblue')
plt.title("cohen's d effect sizes between models")
plt.xlabel("model comparisons")
plt.ylabel("effect size (cohen's d)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# ================================
# plot qq plots of model residuals
# ================================

from scipy.stats import probplot

# create a dictionary of model residuals for iteration
model_residuals = {
    "lstm": residuals_lstm.flatten(),
    "gru": residuals_gru.flatten(),
    "bilstm": residuals_bilstm.flatten(),
    "lr": residuals_lr.flatten(),
    "rf": residuals_rf.flatten()
}

# set figure size and layout
plt.figure(figsize=(15, 10))
n_models = len(model_residuals)
rows = (n_models + 2) // 3  # maximum 3 subplots per row

# iterate over model residuals to plot qq plots
for i, (model_name, residuals) in enumerate(model_residuals.items(), 1):
    plt.subplot(rows, 3, i)  # create subplot
    probplot(residuals, dist="norm", plot=plt)
    plt.title(f'qq plot of {model_name} residuals')

# adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# ================================
# interpretability analysis shap 
# involved models: lstm / gru / bilstm / linear regression / random forest
# ================================

import shap  # note: pay attention to kernel and version compatibility issues
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------
# step 0: flatten model input to 2d for shap
# ---------------------------------------------------

# in the main code, features shape is (n, 60, 6)
# need to flatten it to (n, 360) so that kernelexplainer can parse it
features_2d = features.reshape(features.shape[0], -1)  # (n, 360)
print("flattened features shape:", features_2d.shape)

# dataset split (x_train, x_val, x_test, ...) remains consistent with original training
# split: first 6550 for training, 6550-6914 for validation, after 6914 for testing
x_train_2d = features_2d[:6550]
x_val_2d   = features_2d[6550:6914]
x_test_2d  = features_2d[6914:]

y_train = labels[:6550]
y_val   = labels[6550:6914]
y_test  = labels[6914:]

# ---------------------------------------------------
# step 1: define prediction functions for each model (lstm / gru / bilstm / linear regression)
# ---------------------------------------------------
def lstm_predict_function(x_2d):
    """
    for 2d input (samples, 360), reshape to 3d (samples, 60, 6),
    then call the trained lstm model for prediction and return a 1d array preds.
    """
    x_3d = x_2d.reshape((x_2d.shape[0], 60, 6))
    preds = regressor.predict(x_3d).flatten()  # use flatten() to keep as a 1d array
    return preds

def gru_predict_function(x_2d):
    """
    reshape 2d input to 3d, call the trained gru model for prediction and return a 1d array preds.
    """
    x_3d = x_2d.reshape((x_2d.shape[0], 60, 6))
    preds = regressorGRU.predict(x_3d).flatten()
    return preds

def bilstm_predict_function(x_2d):
    """
    reshape 2d input to 3d, call the trained bilstm model for prediction and return a 1d array preds.
    """
    x_3d = x_2d.reshape((x_2d.shape[0], 60, 6))
    preds = regressorBiLSTM.predict(x_3d).flatten()
    return preds

def linear_regression_predict_function(x_2d):
    """
    directly use the trained linear regression model for prediction and return a 1d array preds.
    """
    preds = lr_model.predict(x_2d).flatten()
    return preds

# ---------------------------------------------------
# random forest uses 2d input, directly use treeexplainer
# ---------------------------------------------------
# rf_model is the trained randomforestregressor

# ---------------------------------------------------
# step 2: prepare background data
# ---------------------------------------------------
# kernelexplainer needs a set of data as the background distribution to approximate shapley values
# background_data = x_train_2d[:500] -- due to lstm progress only 28% in one day, need to change idea or it won't finish in time
background_data = shap.sample(x_train_2d, 60)  # 60 samples, took 12 hours to complete

# ---------------------------------------------------
# step 3: create explainers for each model
# ---------------------------------------------------
# for deep learning models (lstm / gru / bilstm), use kernelexplainer
# for linear models (linear regression), use linearexplainer
# for tree-based models (random forest), use treeexplainer

# lstm explainer
explainer_lstm = shap.KernelExplainer(lstm_predict_function, data=background_data, link='identity')

# gru explainer
explainer_gru = shap.KernelExplainer(gru_predict_function, data=background_data, link='identity')

# bilstm explainer
explainer_bilstm = shap.KernelExplainer(bilstm_predict_function, data=background_data, link='identity')

# linear regression explainer
explainer_lr = shap.LinearExplainer(lr_model, x_train_2d, feature_perturbation="interventional")

# random forest explainer
explainer_rf = shap.TreeExplainer(rf_model)

# ---------------------------------------------------
# step 4: select test set samples to explain
# ---------------------------------------------------
num_explain = 60  # due to time, only explain the first 60 test samples
to_explain_2d = x_test_2d[:num_explain]

# ---------------------------------------------------
# step 5: compute shap values
# ---------------------------------------------------
print("computing shap values for lstm (kernelexplainer) ...")
shap_values_lstm = explainer_lstm.shap_values(to_explain_2d)

print("computing shap values for gru (kernelexplainer) ...")
shap_values_gru = explainer_gru.shap_values(to_explain_2d)

print("computing shap values for bilstm (kernelexplainer) ...")
shap_values_bilstm = explainer_bilstm.shap_values(to_explain_2d)

print("computing shap values for linear regression (linearexplainer) ...")
shap_values_lr = explainer_lr.shap_values(to_explain_2d)

print("computing shap values for random forest (treeexplainer) ...")
shap_values_rf = explainer_rf.shap_values(to_explain_2d)

# ---------------------------------------------------
# step 6: name features for batch differentiation
# ---------------------------------------------------
feature_names = []
time_steps = 60
feat_list = ["Open", "Close", "High", "Low", "Volume", "AdjClose"]
for t in range(time_steps):
    for f in feat_list:
        feature_names.append(f"t{t}_{f}")

# ---------------------------------------------------
# step 7: visualization
# ---------------------------------------------------

print("\nshap summary plot for lstm:")
shap.summary_plot(
    shap_values_lstm, 
    to_explain_2d, 
    feature_names=feature_names, 
    plot_type='bar'  # here 'bar' is more intuitive than 'dot'
)

# gru
print("\nshap summary plot for gru:")
shap.summary_plot(
    shap_values_gru, 
    to_explain_2d, 
    feature_names=feature_names, 
    plot_type='bar'
)

# bilstm
print("\nshap summary plot for bilstm:")
shap.summary_plot(
    shap_values_bilstm, 
    to_explain_2d, 
    feature_names=feature_names, 
    plot_type='bar'
)

# linear regression
print("\nshap summary plot for linear regression:")
shap.summary_plot(
    shap_values_lr, 
    to_explain_2d, 
    feature_names=feature_names, 
    plot_type='bar'
)

# random forest
print("\nshap summary plot for random forest:")
shap.summary_plot(
    shap_values_rf, 
    x_test_2d[:num_explain], 
    feature_names=feature_names, 
    plot_type='bar'
)

# ---------------------------------------------------
# step 8: visualization (force_plot) - local explanation for a single sample
# ---------------------------------------------------
sample_idx = 0  # choose the first sample

print(f"\nshap force plot for sample index: {sample_idx}")

# lstm force plot
print("lstm force plot:")
shap.force_plot(
    explainer_lstm.expected_value, 
    shap_values_lstm[sample_idx, :], 
    to_explain_2d[sample_idx, :],
    feature_names=feature_names,
    matplotlib=True
)
plt.show()

# gru force plot
print("gru force plot:")
shap.force_plot(
    explainer_gru.expected_value, 
    shap_values_gru[sample_idx, :], 
    to_explain_2d[sample_idx, :],
    feature_names=feature_names,
    matplotlib=True
)
plt.show()

# bilstm force plot
print("bilstm force plot:")
shap.force_plot(
    explainer_bilstm.expected_value, 
    shap_values_bilstm[sample_idx, :], 
    to_explain_2d[sample_idx, :],
    feature_names=feature_names,
    matplotlib=True
)
plt.show()

# linear regression force plot
print("linear regression force plot:")
shap.force_plot(
    explainer_lr.expected_value, 
    shap_values_lr[sample_idx, :], 
    to_explain_2d[sample_idx, :],
    feature_names=feature_names,
    matplotlib=True
)
plt.show()

# random forest force plot
print("random forest force plot:")
shap.force_plot(
    explainer_rf.expected_value, 
    shap_values_rf[sample_idx, :], 
    x_test_2d[sample_idx, :],
    feature_names=feature_names,
    matplotlib=True
)
plt.show()

# ---------------------------------------------------
# step 9: visualization of "time step vs. feature" shap values for deep analysis
# ---------------------------------------------------
def visualize_time_step_shap(shap_value_1d, sample_idx=0):
    """
    shap_value_1d: shap value vector for a single sample (length=360), from shap_values_lstm[sample_idx, :] etc.
    sample_idx: used for printing prompt information
    """
    # reshape back to (60, 6)
    shap_2d = shap_value_1d.reshape(60, 6)
    plt.figure(figsize=(10, 6))
    plt.imshow(shap_2d, cmap='RdBu', aspect='auto')
    plt.colorbar(label='shap value')
    plt.xlabel("feature index (0:open,1:close,2:high,3:low,4:volume,5:adjclose)")
    plt.ylabel("time step (0 ~ 59)")
    plt.title(f"sample {sample_idx} timestep-feature shap heatmap")
    plt.show()

# visualize shap heatmap for each model for sample 0
print("\nshap heatmap for lstm:")
visualize_time_step_shap(shap_values_lstm[sample_idx, :], sample_idx=sample_idx)

print("\nshap heatmap for gru:")
visualize_time_step_shap(shap_values_gru[sample_idx, :], sample_idx=sample_idx)

print("\nshap heatmap for bilstm:")
visualize_time_step_shap(shap_values_bilstm[sample_idx, :], sample_idx=sample_idx)

print("\nshap heatmap for linear regression:")
visualize_time_step_shap(shap_values_lr[sample_idx, :], sample_idx=sample_idx)

# for random forest, the features are 2d so can be directly visualized
def visualize_tree_shap(shap_values, sample_idx=0):
    """
    shap_values: shap values from the random forest model
    sample_idx: sample index
    """
    shap_1d = shap_values[sample_idx, :]
    shap_2d = shap_1d.reshape(60, 6)
    plt.figure(figsize=(10, 6))
    plt.imshow(shap_2d, cmap='RdBu', aspect='auto')
    plt.colorbar(label='shap value')
    plt.xlabel("feature index (0:open,1:close,2:high,3:low,4:volume,5:adjclose)")
    plt.ylabel("time step (0 ~ 59)")
    plt.title(f"sample {sample_idx} timestep-feature shap heatmap for random forest")
    plt.show()

print("\nshap heatmap for random forest:")
visualize_tree_shap(shap_values_rf, sample_idx=sample_idx)

print("all shap analysis completed")
```
